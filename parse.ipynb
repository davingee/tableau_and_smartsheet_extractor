```python
from cbps_schema import CBPS_SCHEMA
import base64
from openai import OpenAI

client = OpenAI()

def parse_cbps_dashboard_png_bytes(png_bytes: bytes) -> dict:
    # You can send an image as a base64-encoded data URL. :contentReference[oaicite:1]{index=1}
    b64 = base64.b64encode(png_bytes).decode("utf-8")
    data_url = f"data:image/png;base64,{b64}"

    # Responses API supports image inputs + Structured Outputs. :contentReference[oaicite:2]{index=2}
    resp = client.responses.create(
        model="gpt-4o-mini",  # pick a vision-capable model you have enabled
        temperature=0,
        instructions=(
            "Extract numbers from the dashboard screenshot into the provided JSON schema. "
            "Only use values visible in the image. Convert numbers with commas to integers. "
            "If unclear or missing, use null. "
            "For activity type keys, use exact labels lowercased with spaces replaced by underscores."
        ),
        input=[{
            "role": "user",
            "content": [
                {"type": "input_text", "text": "Return JSON matching the schema."},
                {"type": "input_image", "image_url": data_url},
            ],
        }],
        text={
            "format": {
                "type": "json_schema",
                **CBPS_SCHEMA,
            }
        },
    )

    # The SDK returns structured text output; depending on SDK version, you may parse:
    # - resp.output_text (string) -> json.loads
    # or access already-parsed JSON if your wrapper provides it.
    import json
    return json.loads(resp.output_text)


img_bytes = await file.read()
data = parse_cbps_dashboard_png_bytes(img_bytes)
```
